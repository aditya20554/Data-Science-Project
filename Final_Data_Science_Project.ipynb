```python
df = df.drop(['weight','payer_code','medical_specialty'], axis = 1)
df = df.replace("?",np.nan)
df = df.dropna()
df.isnull().sum()
df["diag_1"] = pd.to_numeric(df["diag_1"], errors="coerce")
df["diag_2"] = pd.to_numeric(df["diag_2"], errors="coerce")
df["diag_3"] = pd.to_numeric(df["diag_3"], errors="coerce")
df = df.dropna()
```

```python
import re
def convert_to_median(row):
    match = re.match(r'\[(\d+)-(\d+)\)', row)
    if match:
        min_value = int(match.group(1))
        max_value = int(match.group(2))
        median = (min_value + max_value) / 2.0
        return median
    else:
        return None  # Handle the case where the regex doesn't match
```

```python
df["age"] = df["age"].apply(convert_to_median)
```

# Validation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, classification_report

def test_confusion_matrix(y_true_test, y_pred_test):
    # Calculate confusion matrix for test set
    cm_test = confusion_matrix(y_true_test, y_pred_test)

    # Visualize confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm_test, annot=True, fmt=".0f")
    plt.xlabel('Predicted Values')
    plt.ylabel('Actual Values')
    plt.title('Confusion Matrix for Test Set')
    plt.show()

    # Other metrics calculation
    precision = precision_score(y_true_test, y_pred_test)
    recall = recall_score(y_true_test, y_pred_test)
    f1 = f1_score(y_true_test, y_pred_test)

    classwise_accuracies = {}
    unique_classes = np.unique(y_true_test)
    for cls in unique_classes:
        mask = y_true_test == cls
        classwise_accuracies[cls] = accuracy_score(y_true_test[mask], y_pred_test[mask])

    weighted_f1 = f1_score(y_true_test, y_pred_test, average='weighted')
    class_report = classification_report(y_true_test, y_pred_test)

    print("Confusion Matrix (Test): \n", cm_test, "\n")
    print("Precision : ", precision)
    print("Recall : ", recall)
    print("F1-score : ", f1, "\n")
    print("Classwise Accuracies: ", classwise_accuracies, "\n")
    print("Weighted F1-score : ", weighted_f1, "\n")
    print("Classification Report : \n", class_report, "\n")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(y_true, y_prob):
    fpr, tpr, thresholds = roc_curve(y_true, y_prob[:, 1])
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='red', lw=1, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
```
